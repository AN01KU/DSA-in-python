            BINARY SEARCH
1.Select middle element, if query number equal to middle element return middle element.
2.If middle element is less than query number , search first half of list.
3.If middle element is more than query number , search second half of list.
4.If no more elements remain, return 1.

```python
def locate_card(cards, query):
    low = 0
    high = len(cards) - 1
    middle = (low + high) // 2

    while low <= high:
        print(f"low: {low}, high: {high}, middle: {middle}")
        if cards[middle] == query:
            return middle
        elif cards[middle] < query:
            high = middle - 1
        else:
            low = middle + 1
        middle = (low + high) // 2
    return -1
```

    Generic Binary Search

1.Come with a condition to determine wether the answer lies before, after or at a given positon.
2.Retrieve midpoint and middle element of list.
3.If it is the answer, return middle position as the answer.
4.If answer lies before it, repeat the search with first half of list.
5.If answer lies after it, repeat the search with second half of list.

```python
def binary_search(low, high, condition):
    """TODO - add docs"""
    while low <= high:
        middle = (low + high) // 2
        result = condition(middle)
        if result == 'found':
            return middle
        elif result == 'left':
            high = middle - 1
        elif result == 'right':
            low = middle + 1
    return -1
```
