    LINEAR SEARCH VS BINARY SEARCH

```python
def locate_card_linear(cards,query):
    position = 0
    while position < len(cards):
        if cards[position] == query:
            return position
        position += 1
    return -1


def test_location(cards, query,mid):
    if cards[mid] == query:
        if mid-1 >= 0 and cards[mid-1] == query:
            return 'left'
        else:
            return 'found'
    elif cards[mid] < query:
        return 'left'
    else:
        return 'right'


def locate_card_binary(cards, query):
    low = 0
    high = len(cards) - 1
    middle = (low + high) // 2

    while low <= high:
        result = test_location(cards, query, middle)
        if result == 'found':            
            return middle
        elif result == 'left':
            high = middle - 1
        elif result == 'right':
            low = middle + 1
        middle = (low + high) // 2
    return -1
```

```python
large_test={
    'input':{
        'cards':list(range(10000000,0,-1)),
        'query': 2
    },
    'output': 9999998
}
```
