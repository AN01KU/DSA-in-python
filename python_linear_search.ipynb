            LINEAR SEARCH
            
1.Create a variable position with value 0.
2.Check if number at index in card is equals query.
3.If it does , then return the position.
4.If not , increment the position by 1 and repeat step 2 to 5 until we reach last index.
5.If number not found , return -1.

```python
test = { 'input': {'cards': [13, 11, 10, 7, 4, 3, 1, 0] , 'query': 7},
         'output': 3 }
```

```python
def locate_card(cards,query):
    position=0  

    while True :
        if cards[position] == query:
            return position
        position += 1
        if position==len(cards):
            return -1
```
